```
import pandas as pd
from matplotlib import pyplot as plt
```

```
data = pd.read_csv('/dbfs/FileStore/tables/dataii.csv')
data.head()
```

```
us = data[data.country == 'United States']
us.head()
```

```
plt.plot(us.year, us.gdpPerCapita)
plt.title('GDP Per Capita Growth in the US')
plt.xlabel('year')
plt.ylabel('GDP Per Capita')
plt.show()
```

```
china = data[data.country == 'China']
china.head()
```

```
plt.plot(us.year, us.gdpPerCapita)
plt.plot(china.year, china.gdpPerCapita)
plt.legend(['United States', 'China'])
plt.xlabel('Year')
plt.ylabel('GDP Per Capita')
plt.show()
```

```
chinagrowth = china.gdpPerCapita / china.gdpPerCapita.iloc[0] * 100
```

```
plt.title('GDP Per Capita Growth (first year equals to 100)')
plt.plot(us.year, usgrowth)
plt.plot(china.year, chinagrowth)
plt.legend(['US', 'China'])
plt.xlabel('Year')
plt.ylabel('GDP Per Capita Growth')
plt.show()
```
