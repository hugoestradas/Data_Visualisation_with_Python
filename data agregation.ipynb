```
import pandas as pd
from matplotlib import pyplot as plt
```

```
data = pd.read_csv('/dbfs/FileStore/tables/obama.csv', parse_dates=['year_month'])
data.head()
```

```
plt.plot(data.year_month, data.approve_percent, 'o', markersize=2, alpha=0.3)
plt.show()
```

```
datamean= data.groupby('year_month').mean()
plt.plot(datamean.index, datamean.approve_percent, 'red')
plt.show()
```

```
datamean= data.groupby('year_month').mean()
plt.plot(datamean.index, datamean.approve_percent, 'red')
plt.plot(data.year_month, data.approve_percent, 'o', markersize=2, alpha=0.3)
plt.show()
```

```
datamedian= data.groupby('year_month').median()
```

```
datamean= data.groupby('year_month').mean()
datamedian= data.groupby('year_month').median()
plt.plot(datamean.index, datamean.approve_percent, 'red')
plt.plot(datamedian.index, datamedian.approve_percent, 'green')
plt.legend(['Mean', 'Median'])
plt.plot(data.year_month, data.approve_percent, 'o', markersize=2, alpha=0.3)
plt.show()
```
